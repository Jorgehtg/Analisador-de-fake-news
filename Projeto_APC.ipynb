```python
from collections import Counter
def clear_string(string: str) -> str:
    string = string.strip()
    if string.startswith('"'):
        string = string[1:]
    if string.endswith('"'):
        string = string[:-2]
    string = string.replace('""', '"')
    return string
def ler_arquivo(treinando_teste):
    titulos = []
    valores = []
    arquivo = open(treinando_teste, 'r')
    linha = arquivo.readlines()
    for i in range(len(linha)):
        line = linha[i]
        frase = clear_string(line[2:])
        valor = int(line[0])
        titulos.append(frase)
        valores.append(valor)
    return valores,titulos
def levenshtein(s1, s2):
    m, n = len(s1), len(s2)
    dp = [[0]*(n+1) for _ in range(m+1)]
    for i in range(m+1): dp[i][0] = i
    for j in range(n+1): dp[0][j] = j
    for i in range(1, m+1):
        for j in range(1, n+1):
            cost = 0 if s1[i-1] == s2[j-1] else 1
            dp[i][j] = min(
                dp[i-1][j] + 1,
                dp[i][j-1] + 1,
                dp[i-1][j-1] + cost
            )
    return dp[m][n]
def knn(treinamento, valor_treinamento, titulo_teste, k):
    distancias = []
    for i in range(len(treinamento)):
        dist = levenshtein(titulo_teste, treinamento[i])
        distancias.append((dist, valor_treinamento[i]))
    distancias.sort(key=lambda x: x[0])
    vizinhos = [label for _, label in distancias[:k]]
    return Counter(vizinhos).most_common(1)[0][0]
def prediçao(treiv, treit, testet, testev, k):
    tp = tn = fp = fn = 0
    for i in range(len(testet)):
        predito = knn(treit, treiv, testet[i], k)
        if testev[i] == 1 and predito == 1:
            tp += 1
        elif testev[i] == 0 and predito == 0:
            tn += 1
        elif testev[i] == 0 and predito == 1:
            fp += 1
        elif testev[i] == 1 and predito == 0:
            fn += 1
    acc = (tp + tn) / (tp + tn + fp + fn)
    precisão = tp / (tp + fp)
    recall = tp / (tp + fn)
    f1sc = 2 * (precisão * recall) / (precisão + recall)
    return acc, precisão, recall, f1sc
treinamento, teste, k = input().split()
k = int(k)
treiv, treit = ler_arquivo(treinamento)
testev, testet = ler_arquivo(teste)
acc, pre, re, f1 = prediçao(treiv, treit, testet, testev, k)
print(f"+-------------------+\n|  Metric   | Value |\n+-------------------+\n| Accuracy  | {acc:.2f}  |\n| Precision | {pre:.2f}  |\n| Recall    | {re:.2f}  |\n| F1-Score  | {f1:.2f}  |\n+-------------------+")
```
